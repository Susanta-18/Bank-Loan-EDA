```python
# Finding some correlation for numerical columns for both target 0 and 1 

ad0_corr=ad0.iloc[0:,2:]
ad1_corr=ad1.iloc[0:,2:]

T0=ad0_corr.corr(method='spearman')
T1=ad1_corr.corr(method='spearman')
```

```python
# Now, plotting the above correlation with heat map
# figure size

def targets_corr(data,title):
    plt.figure(figsize=(15, 10))
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 50

# heatmap with a color map of choice


    sns.heatmap(data, cmap="RdYlGn",annot=False)

    plt.title(title)
    plt.yticks(rotation=0)
    plt.show()
```

```python
# For Target 0

targets_corr(data=T0,title='Correlation for target 0')
```

```python
non_defaulter_corr = T0.corr()
round(non_defaulter_corr, 2)

corr_list = non_defaulter_corr.unstack()
```

```python
# Listing the correlations in pair sorted in descending order
corr_list.sort_values(ascending=False).drop_duplicates().head(11)
```

```python
targets_corr(data=T1,title='Correlation for target 1')
```

```python
defaulter_corr = T1.corr()
round(defaulter_corr, 2)

corr_list_1 = defaulter_corr.unstack()
```
